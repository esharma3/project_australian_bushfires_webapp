```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd
import plotly.offline as py
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import plotly.express as px
import re

df_2019_2020_fires = pd.read_csv(".\impact_data\df_2019_2020_fires.csv")
# df_2019_2020_fires = df_2019_2020_fires.drop(columns=['notes'])
df_2019_2020_fires = df_2019_2020_fires.drop([df_2019_2020_fires.index[8]])

fig = make_subplots(
    rows=4, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05,
    specs=[[{"type": "table"}],
           [{"type": "scatter"}],
           [{"type": "scatter"}],
           [{"type": "scatter"}]]
)

fig.add_trace(
    go.Scatter(
        x=df_2019_2020_fires["state"],
        y=df_2019_2020_fires["homes_destroyed"],
        mode="lines",
        name="2019-2020 Bushfires Homes Destroyed By State",
    ),
    row=4, col=1
)

fig.add_trace(
    go.Scatter(
        x=df_2019_2020_fires["state"],
        y=df_2019_2020_fires["acres_burned"],
        mode="lines",
        name="2019-2020 Bushfires Acres Burned By State"
    ),
    row=3, col=1
)

fig.add_trace(
    go.Scatter(
        x=df_2019_2020_fires["state"],
        y=df_2019_2020_fires["human_fatalities"],
        mode='lines',
        name="2019-2020 Bushfires Human Fatalities By State"
    ),
    row=2, col=1
)

fig.add_trace(
    go.Table(
        header=dict(
            values=["State", "Human Fatalities", "Homes Destroyed",
                    "Hectacres Burned", "Acres Burned"],
            font=dict(size=10),
            align="center"
        ),
        cells=dict(
            values=[df_2019_2020_fires[k].tolist() for k in df_2019_2020_fires.columns[1:]],
            align = "center")
    ),
    row=1, col=1
)

fig.update_layout(
    height=2500,
    width=800,
    legend=dict(
        x=0,
        y=1,
        traceorder="reversed",
        font=dict(
            family="sans-serif",
            size=6,
            color="black"
        ),
        bgcolor="White",
        bordercolor="Black",
        borderwidth=1
    )
)

fig.update_layout(legend=dict(x=1.011, y=.3))
fig.update_layout(
    title={
        'text': "Australian Bushfire Impact 2019-2020",
        'y':.77,
        'x':0.4,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20),
    paper_bgcolor="LightGrey",
)

###############
# fig.show()
# py.plot(fig, filename='2019_2020 Fire Impact.html')
```
