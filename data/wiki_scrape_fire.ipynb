```python
import requests 
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import html5lib
import pprint
import os
import codecs
import re


url = 'https://en.wikipedia.org/wiki/List_of_major_bushfires_in_Australia' 
r = requests.get(url)
soup = BeautifulSoup(r.content)
# type(soup)

table=soup.find('table', {'class', 'wikitable sortable'})

headers= [header.text for header in table.find_all('th')]
headers.remove('Properties damaged\n')
headers.remove('Area burned (approx.)\n')
headers


rows = []
for row in table.find_all('tr'):
    rows.append([val.text.encode('utf8') for val in row.find_all('td')])
    # pprint.pprint(rows)
    
df_hist_fires = pd.DataFrame(rows, columns=headers)

#create header titles, decode from UTF8
df_hist_fires.columns = ['date', 'name', 'state', 'hectacres_burned', 'acres_burned', 'fatalities', 'homes_lost', 'other_lost', 'other_damage', 'notes']
df_hist_fires['date'] = df_hist_fires['date'].str.decode('utf-8')
df_hist_fires['name'] = df_hist_fires['name'].str.decode('utf-8')
df_hist_fires['state'] = df_hist_fires['state'].str.decode('utf-8')
df_hist_fires['hectacres_burned'] = df_hist_fires['hectacres_burned'].str.decode('utf-8')
df_hist_fires['acres_burned'] = df_hist_fires['acres_burned'].str.decode('utf-8')
df_hist_fires['fatalities'] = df_hist_fires['fatalities'].str.decode('utf-8')
df_hist_fires['homes_lost'] = df_hist_fires['homes_lost'].str.decode('utf-8')
df_hist_fires['other_lost'] = df_hist_fires['other_lost'].str.decode('utf-8')
df_hist_fires['other_damage'] = df_hist_fires['other_damage'].str.decode('utf-8')

#removing decode values
df_hist_fires.drop(["notes"], axis = 1, inplace = True)
df_hist_fires = df_hist_fires.replace(r'\\n',' ', regex=True)
df_hist_fires = df_hist_fires.replace('\n',' ', regex=True)

#clean date, fatalities and homes lost
df_hist_fires.name.str.split(expand=True,)
df_hist_fires['year'] = df_hist_fires['date'].str.extract('(\d\d\d\d)', expand=True)
df_hist_fires['human_fatalities'] = df_hist_fires['fatalities'].str.extract('(\d\d)', expand=True)

df_hist_fires['homes_destroyed'] = df_hist_fires['homes_lost'].str.split(".").str[-1].str.replace(',', '')
df_hist_fires['homes_destroyed'] = df_hist_fires['homes_destroyed'].str.replace(',', '')
df_hist_fires['homes_destroyed'] = df_hist_fires['homes_destroyed'].str.replace('approx.', '')
df_hist_fires['homes_destroyed'] = df_hist_fires['homes_destroyed'].str.extract(r'(\d+)', expand=True)

df_hist_fires['other_lost'] = df_hist_fires['other_lost'].str.replace(',', '')
df_hist_fires['other_lost'] = df_hist_fires['other_lost'].str.extract(r'(\d+)', expand=True)

df_hist_fires['human_fatalities'] = df_hist_fires['fatalities'].str.split(".").str[-1]
df_hist_fires['human_fatalities'] = df_hist_fires['human_fatalities'].str.extract(r'(\d+)', expand=True)

#dropping unclean columns
df_hist_fires.drop(["date","fatalities","homes_lost", "other_lost"], axis = 1, inplace = True)
df_hist_fires = df_hist_fires.drop([0,1])
df_hist_fires.dropna()


df_hist_fires.head(1)


# df_hist_fires.to_csv(r'C:\Users\james\Desktop\project_australian_bushfires_webapp\data\impact_data\df_hist_fires.csv')












```
