```python
import requests 
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import html5lib
import pprint
import os
import codecs
import re


url = 'https://en.wikipedia.org/wiki/List_of_major_bushfires_in_Australia' 
r = requests.get(url)
soup = BeautifulSoup(r.content)
# type(soup)

table=soup.find('table', {'class', 'wikitable sortable'})

headers= [header.text for header in table.find_all('th')]
headers.remove('Properties damaged\n')
headers.remove('Area burned (approx.)\n')
headers


rows = []
for row in table.find_all('tr'):
    rows.append([val.text.encode('utf8') for val in row.find_all('td')])
    # pprint.pprint(rows)
    
df_hist_fires = pd.DataFrame(rows, columns=headers)

#create header titles, decode from UTF8
df_hist_fires.columns = ['Date', 'Name', 'State', 'Hectacres Burned', 'Acres Burned', 'Fatalities', 'Homes Lost', 'Other Lost', 'Other Damage', 'Notes']
df_hist_fires['Date'] = df_hist_fires['Date'].str.decode('utf-8')
df_hist_fires['Name'] = df_hist_fires['Name'].str.decode('utf-8')
df_hist_fires['State'] = df_hist_fires['State'].str.decode('utf-8')
df_hist_fires['Hectacres Burned'] = df_hist_fires['Hectacres Burned'].str.decode('utf-8')
df_hist_fires['Acres Burned'] = df_hist_fires['Acres Burned'].str.decode('utf-8')
df_hist_fires['Fatalities'] = df_hist_fires['Fatalities'].str.decode('utf-8')
df_hist_fires['Homes Lost'] = df_hist_fires['Homes Lost'].str.decode('utf-8')
df_hist_fires['Other Lost'] = df_hist_fires['Other Lost'].str.decode('utf-8')
df_hist_fires['Other Damage'] = df_hist_fires['Other Damage'].str.decode('utf-8')

#removing decode values
df_hist_fires.drop(["Notes"], axis = 1, inplace = True)
df_hist_fires = df_hist_fires.replace(r'\\n',' ', regex=True)
df_hist_fires = df_hist_fires.replace('\n',' ', regex=True)

#clean date, fatalities and homes lost
df_hist_fires.Name.str.split(expand=True,)
df_hist_fires['Year'] = df_hist_fires['Date'].str.extract('(\d\d\d\d)', expand=True)
df_hist_fires['Human Fatalities'] = df_hist_fires['Fatalities'].str.extract('(\d\d)', expand=True)

df_hist_fires['Homes Destroyed'] = df_hist_fires['Homes Lost'].str.split(".").str[-1].str.replace(',', '')
df_hist_fires['Homes Destroyed'] = df_hist_fires['Homes Destroyed'].str.replace(',', '')
df_hist_fires['Homes Destroyed'] = df_hist_fires['Homes Destroyed'].str.replace('approx.', '')
df_hist_fires['Homes Destroyed'] = df_hist_fires['Homes Destroyed'].str.extract(r'(\d+)', expand=True)

df_hist_fires['Other Lost'] = df_hist_fires['Other Lost'].str.replace(',', '')
df_hist_fires['Other Structures Destroyed'] = df_hist_fires['Other Lost'].str.extract(r'(\d+)', expand=True)

df_hist_fires['Human Fatalities'] = df_hist_fires['Fatalities'].str.split(".").str[-1]
df_hist_fires['Human Fatalities'] = df_hist_fires['Human Fatalities'].str.extract(r'(\d+)', expand=True)

#dropping unclean columns
df_hist_fires.drop(["Date","Fatalities","Homes Lost", "Other Lost"], axis = 1, inplace = True)

df_hist_fires.dropna()


df_hist_fires.head(10)


df_hist_fires.to_csv(r'df_hist_fires.csv')












```
