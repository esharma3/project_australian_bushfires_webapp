```python
#Australian GDP scrape
import requests 
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import html5lib
import pprint
import os
import codecs
import re


url = 'https://en.m.wikipedia.org/wiki/Economy_of_Australia' 
r = requests.get(url)
soup = BeautifulSoup(r.content)
# type(soup)

table=soup.find('table', {'class', 'wikitable'})

headers= [header.text for header in table.find_all('th')]
# headers.remove('Properties damaged\n')
# headers.remove('Area burned (approx.)\n')
# headers


rows = []
for row in table.find_all('tr'):
    rows.append([val.text.encode('utf8') for val in row.find_all('td')])
    #pprint.pprint(rows)
df_aus_econs = pd.DataFrame(rows, columns=headers)

#create header titles, decode from UTF8
df_aus_econs.columns = ['year', 'gdp_us_bn', 'gdp_capita_us', 'gdp_growth_perc', 'inflation_rate_perc', 'unemployment_perc', 'gov_debt_perc']
df_aus_econs['year'] = df_aus_econs['year'].str.decode('utf-8')
df_aus_econs['gdp_us_bn'] = df_aus_econs['gdp_us_bn'].str.decode('utf-8')
df_aus_econs['gdp_capita_us'] = df_aus_econs['gdp_capita_us'].str.decode('utf-8')
df_aus_econs['gdp_growth_perc'] = df_aus_econs['gdp_growth_perc'].str.decode('utf-8')
df_aus_econs['inflation_rate_perc'] = df_aus_econs['inflation_rate_perc'].str.decode('utf-8')
df_aus_econs['unemployment_perc'] = df_aus_econs['unemployment_perc'].str.decode('utf-8')
df_aus_econs['gov_debt_perc'] = df_aus_econs['gov_debt_perc'].str.decode('utf-8')

# remove decode values
df_aus_econs = df_aus_econs.replace(r'\\n',' ', regex=True)
df_aus_econs = df_aus_econs.replace('\n',' ', regex=True)

#remove % values and additional punctuations 
df_aus_econs['gdp_growth_perc'] = df_aus_econs['gdp_growth_perc'].str.replace('%', '')
df_aus_econs['gdp_capita_us'] = df_aus_econs['gdp_capita_us'].str.replace(',', '')
df_aus_econs['inflation_rate_perc'] = df_aus_econs['inflation_rate_perc'].str.replace('%', '')
df_aus_econs['unemployment_perc'] = df_aus_econs['unemployment_perc'].str.replace('%', '')
df_aus_econs['gov_debt_perc'] = df_aus_econs['gov_debt_perc'].str.replace('%', '')
df_aus_econs['gov_debt_perc'] = df_aus_econs['gov_debt_perc'].str.replace('n/a', '')
df_aus_econs.dropna()

# df_aus_econs.head(10)

df_aus_econs.to_csv(r'df_aus_econs.csv')
```
