```python
rows = table.findAll("tr")

# strip the header from rows
headers = rows[0]
header_text = []

# add the header text to array
for th in headers.findAll("th"):
    header_text.append(th.text)

row_text_array = []

# loop through rows and add row text to array
for row in rows[1:]:
    row_text = []
    for row_element in row.findAll(["th", "td"]):
        row_text.append(row_element.text.replace("\n", "").strip())
    row_text_array.append(row_text)

# output csv
with open("data/climate_anomalies.csv", "w") as f:
    wr = csv.writer(f)
    wr.writerow(header_text)
    for row_text_single in row_text_array:
        wr.writerow(row_text_single)
```

```python
climate_anomalies_df = pd.read_csv(
    "data/climate_anomalies.csv", skiprows=1, encoding="ISO-8859-1"
).rename(
    columns={
        "Unnamed: 0": "year",
        "Annual mean temperature anomaly": "annual_mean_temperature_anomaly",
        "Rank": "temp_rank",
        "Fraction of Australia in highest decile": "fraction_of_australia_in_highest_decile",
        "Fraction of Australia in lowest decile": "fraction_of_australia_in_lowest_decile",
        "Annual mean rainfall": "annual_mean_rainfall",
        "Rank.1": "rainfall_rank",
        "Fraction of Australia in highest decile.1": "fraction_of_australia_in_highest_decile",
        "Fraction of Australia in lowest decile.1": "fraction_of_australia_in_lowest_decile",
        "Annual mean SST anomaly": "annual_mean_sea_surface_temp_anomaly",
        "Rank.2": "sst_rank",
    }
)
```

```python
climate_anomalies_df.head()
```

```python
climate_anomalies_df["annual_mean_temperature_anomaly"] = (
    climate_anomalies_df["annual_mean_temperature_anomaly"].str[1:5].astype(float)
)

climate_anomalies_df["annual_mean_sea_surface_temp_anomaly"] = (
    climate_anomalies_df["annual_mean_sea_surface_temp_anomaly"].str[1:5].astype(float)
)

climate_anomalies_df["annual_mean_rainfall"] = (
    climate_anomalies_df["annual_mean_rainfall"].str[:-2].astype(float)
)
```

```python
# visualizing Monthly Minimum Median Wages to Average Rents for different apartment sizes

trace1 = go.Scatter(x=climate_anomalies_df["year"],
                  y=climate_anomalies_df['annual_mean_temperature_anomaly'], 
#                   width=0.2,
                  marker=dict(color="red", line_color='red'),
                  name='Annual mean change in temperature (C)')

trace2 = go.Scatter(x=climate_anomalies_df["year"],
                  y=climate_anomalies_df['annual_mean_sea_surface_temp_anomaly'], 
#                   width=0.2,
                  marker=dict(color="black", line_color='black'),
                  name='Annual mean SST anomaly')

data = [trace1, trace2]

layout = dict(title="<b>Australia - Annual mean temperature (C) anomaly</b>",
              showlegend=True,
              legend_orientation="v",
              autosize=True,
#               width=1010,
#               height=600,
              xaxis=dict(showgrid=False),
              yaxis=dict(title="Annual mean change in temperature (C)", 
                         showgrid=True, 
                         autorange=True,
                         zeroline=False,
                         showline=False,
                         ticks='',
                         showticklabels=True),
              plot_bgcolor='rgba(0,0,0,0)',
              paper_bgcolor='rgba(0,0,0,0)',
          )

fig=dict(data=data, layout=layout)

iplot(fig)
```
